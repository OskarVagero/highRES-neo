* magasinkapasitet_og_reguleringsgrad_vannkraftverk.xlsx was acquired from personal correspondence between Oskar Vågerö (ITS) and Carl Andreas Veie <cave@nve.no> (NVE). 
    * Hei. Det er ikke helt åpenbart hvordan man skal dele inn kraftverk som magasinkraftverk og elvekraftverk, og inndelingen avhenger ofte av hvordan data skal brukes. Noen kraftverk har selv ikke magasiner, men utnytter regulert vann fra oppstrøms magasiner. For eksempel har Vamma kraftverk, som du nevner, 3585 mill. m3 med magasinkapasitet oppstrøms kraftverket, selv om dette kraftverket ikke har noe eget magasin. Derfor kan det være nyttig å dele inn kraftverk etter hvor mye oppstrøms magasinkapasitet de har. Vi har definert en parameter kalt global reguleringsgrad, også kalt magasinprosent , som er sum oppstrøms magasinkapasitet (i mill. m^3) delt på sum midlere årstilsig (i mill. m^3) til kraftverket. I denne sammenhengen har vi definert magasinkraftverk som kraftverk hvor global reguleringsgrad er høyere enn 0,5, altså at sum oppstrøms magasinkapasitet utgjør mer enn halvparten av midlere årstilsig til kraftverket. Det er ikke gitt at denne definisjonen passer til alle formål, men den sier noe viktig om hvor stor evne man har til å regulere tilsiget som kraftverket utnytter. Si ifra om dere har noen spørsmål til oversendte data eller om dere har behov for annen vannkraftdata.
