```python
# Show the 'extremes', i.e. the technology capacities for the lowest and highest solar costs that are sampled compared to the baselinelabelnames = ['lowest solar costs', 'baseline', 'highest solar costs']low_data = simulated_data.iloc[id_lowest_solarcosts]high_data = simulated_data.iloc[id_highest_solarcosts]baseline = baseline_data.iloc[0]labels = [low_data, baseline, high_data]ExampleData = pd.DataFrame(labels)ExampleData.pop('Total system costs')ExampleData.pop('Solar costs')ExampleData.pop('Nuclear')width = 0.5fig, ax = plt.subplots()# Write the respective values inside the bars (https://stackoverflow.com/questions/32950564/y-values-on-each-stacked-bar-using-matplotlib)def autolabel(rects, values):# Attach some text labels.for (rect, value) in zip(rects, values):ax.text(rect.get_x() + rect.get_width() / 2.,rect.get_y() + rect.get_height() / 2.,'%d'%value,ha = 'center',va = 'center',color = 'white')# Stack the different technology capacitiesbot = 0 # specify how they are to be stacked and do this recursivelyfor j in technologies:r = ax.bar(labelnames, ExampleData[j], width, label=j, bottom = bot, color = technologies[j][1])autolabel(r, ExampleData[j])bot += ExampleData[j]ax.set_ylabel('Capacity in GW')ax.set_title('Detailed technology capacities in different samples')ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')plt.show()# Show the 'extremes', i.e. the technology capacities for the lowest and highest solar costs that are sampled compared to the baseline
labelnames = ['lowest solar costs', 'baseline', 'highest solar costs']
low_data = simulated_data.iloc[id_lowest_solarcosts]
high_data = simulated_data.iloc[id_highest_solarcosts]
baseline = baseline_data.iloc[0]
labels = [low_data, baseline, high_data]
ExampleData = pd.DataFrame(labels)
ExampleData.pop('Total system costs')
ExampleData.pop('Solar costs')
ExampleData.pop('Nuclear')
width = 0.5      
fig, ax = plt.subplots()
# Write the respective values inside the bars (https://stackoverflow.com/questions/32950564/y-values-on-each-stacked-bar-using-matplotlib)
def autolabel(rects, values):
    # Attach some text labels.
    for (rect, value) in zip(rects, values):
        ax.text(rect.get_x() + rect.get_width() / 2.,
                rect.get_y() + rect.get_height() / 2.,
                '%d'%value,
                ha = 'center',
                va = 'center',
                color = 'white')
# Stack the different technology capacities
bot = 0 # specify how they are to be stacked and do this recursively
for j in technologies:
    r = ax.bar(labelnames, ExampleData[j], width, label=j, bottom = bot, color = technologies[j][1])
    autolabel(r, ExampleData[j])
    bot += ExampleData[j]
ax.set_ylabel('Capacity in GW')
ax.set_title('Detailed technology capacities in different samples')
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')   
plt.show()
```

## Installed capacity

```python
df_cap = pd.read_sql_query("SELECT * from var_tot_pcap_z", con)
df_cap = df_cap.drop(columns=["lo", "up", "marginal"])
```

```python
df_cap
```

```python
# We rename the columns to something which is a bit easier to understand.
df_cap = df_cap.rename(
    columns={
        "z": "country",
        "g": "technology",
        "level" : "installed capacity [?]"
    }
)
```

```python
# We also rename the technology names to something a bit more tidy.
df_cap.technology = df_cap.technology.replace(
    {
        "NaturalgasCCGTwithCCSnewOT": "Gas with CCS",
        "HydroRes" : "Hydro reservoir",
        "HydroRoR" : "Hydro run-of-river",
        "Windonshore" : "Onshore wind",
        "Windoffshore" : "Offshore wind",
        "NaturalgasOCGTnew" : "Gas without CCS"
    }
)
```

```python
df_cap.set_index(["country", "technology"]).unstack().loc[:,"installed capacity [?]"]
```

```python
# For all countries
df_cap.set_index(
    ["country",
     "technology"]
).unstack(
).loc[
    :,"installed capacity [?]"
].plot.bar(
    figsize=(20,10),
    title="Installed capacities [GW]",
    color=["#CC79A7","#d35050","#08ad97","#4adbc8","#ff9000","#6895dd","#235ebc","#f9d002"],
    stacked=True,
    ylim=(0,220),
)
```

```python
# For a selected number of countries
df_cap[df_cap["country"].isin(["PT","GR","DK","FI"])].set_index(["country","technology"]).unstack().loc[:,"installed capacity [?]"].plot.bar(
    figsize=(20,10),
    title="Installed capacities [GW]",
    color=["#CC79A7","#000000","#0072B2","#56B4E9","#D55E00","#E69F00","#009E73","#F0E442"],
    stacked=True,
    ylim=(0,100),
)
```

## Generation

```python
df_gen = pd.read_sql_query("SELECT * from var_gen", con)
df_gen = df_gen.drop(columns=["lo", "up", "marginal"])
```

```python
# We rename the columns to something which is a bit easier to understand.
df_gen = df_gen.rename(
    columns={
        "z": "country",
        "g": "technology",
        "h": "hour",
        "level" : "Electricity generated"
    }
)
```

```python
df_gen.technology = df_gen.technology.replace(
    {
        "NaturalgasCCGTwithCCSnewOT": "Gas with CCS",
        "HydroRes" : "Hydro reservoir",
        "HydroRoR" : "Hydro run-of-river",
        "Windonshore" : "Onshore wind",
        "Windoffshore" : "Offshore wind",
        "NaturalgasOCGTnew" : "Gas without CCS"
    }
)
```

```python
df_gen
```

```python
df_gen[df_gen["country"].isin(["SE","DK","NO"])].set_index(
    ["hour",
     "country"]
).groupby(
    ["country",
     "hour"]
).sum(
).unstack(
).loc[
    :,"Electricity generated"
].transpose(
).plot(
)
```

### Demand

This is purely based on input data, and is not a variable in the model.

```python
df_dem = pd.read_sql_query("SELECT * from demand", con)
```

```python
# We rename the columns to something which is a bit easier to understand.
df_dem = df_dem.rename(
    columns={
        "z": "country",
        "h": "hour",
        "value" : "Electricity demand [?]"
    }
)
```

```python
df_dem[df_dem["country"].isin(["SE","DK","NO"])].set_index(
    ["hour",
     "country"]
).groupby(
    ["hour",
     "country"]
).sum(
).unstack(
).loc[
    :,"Electricity demand [?]"
].plot()
```

### Emissions

```python
df_emi = pd.read_sql_query("SELECT * from o_emissions", con)
```

```python
# We rename the columns to something which is a bit easier to understand.
df_emi = df_emi.rename(
    columns={
        "z": "country",
        "h": "hour",
        "non_vre" : "technology",
        "value" : "emissions [?]"
    }
)
```

```python
df_emi.technology = df_emi.technology.replace(
    {
        "NaturalgasCCGTwithCCSnewOT": "Gas with CCS",
        "HydroRes" : "Hydro reservoir",
        "HydroRoR" : "Hydro run-of-river",
        "Windonshore" : "Onshore wind",
        "Windoffshore" : "Offshore wind",
        "NaturalgasOCGTnew" : "Gas without CCS"
    }
)
```

```python
df_emi
```

```python
df_emi.set_index(
     ["country",
      "technology"]
).groupby(
     ["country",
      "technology"]
).sum(
).unstack(
).loc[
    :,"emissions [?]"
].sort_values(
    "Gas without CCS",
    ascending=False
).plot.bar(
    stacked=True,
    figsize=(20,10),
    color = ["#E69F00", "#000000", "#009E73"]
)
```

### Transmission capacity

```python
df_trans_cap = pd.read_sql_query("SELECT * from var_trans_pcap", con)
df_trans_cap = df_trans_cap.drop(columns=["lo", "up", "marginal"])
```

```python
# We rename the columns to something which is a bit easier to understand.
df_trans_cap = df_trans_cap.rename(
    columns={
        "z": "country 1",
        "z_alias" : "country 2",
        "trans" : "type",
        "level" : "capacity [?]"
    }
)
```

```python
df_trans_cap
```

```python
df_trans_cap.set_index(
    "country 1"
).groupby(
    "country 1"
).sum(
).sort_values(
    "capacity [?]",
    ascending=False
).plot.bar(
    figsize=(14,8),
    color = "#009E73"
)
```
