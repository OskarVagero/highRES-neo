## Capacity factors

Here we need to use the following parameters:

* matrix (N x S - xr.DataArray or sp.sparse.csr_matrix or None)  If given, it is used to aggregate the grid cells to buses. N is the number of buses, S the number of spatial coordinates, in the order of cutout.grid.
    - Since we want to aggregate grid cells to shapes (countries), which are called buses in atlite, this is necessary.
* per_unit (boolean)  Returns the time-series in per-unit units, instead of in MW (defaults to False).
    - Since we want capacity factors and not capacities, we have to use per_unit. Per-unit units means the factor between 0 and 1 of the total capacity. Since we are using this relative measure, the total capacity does not matter and the capacity_matrix should therefor only be used to exclude cells.
* index (pd.Index)  Index of Buses.
    - This is the index of the shapes.

```python
capacity_factor_solar = cutout.pv(
    panel=panel,
    orientation=orientation,
    matrix=capacity_matrix_solar,
    # capacity_factor=True, # does not do anything here
    per_unit=True,
    index=europe.index
)

capacity_factor_solar
```

### Calculating Norwegian fractions

```python
hydro_gen = pd.read_csv(
    'Vannkraftverk.csv', 
    encoding = "latin", 
    skiprows=2, 
    sep=";", 
    decimal = ",",
    index_col = False
)
```

```python
hydro_gen
```

```python
hydro_fractions = hydro_gen.rename(
    columns={
        'Midl. rsproduksjon (ref. 1981-2010) [GWh]' : 'average_generation_GWh',
        'Maks ytelse [MW]' : 'installed_capacity_MW'
    }
).groupby(
    'Fylke'
).sum(
).assign(
    share_gen = lambda x : x.average_generation_GWh/x.average_generation_GWh.sum(),
    share_cap = lambda x : x.installed_capacity_MW/x.installed_capacity_MW.sum(),
    NUTS3_ID = ['NO092','NO020','NO0A3','NO071','NO081','NO0A1','NO074','NO060','NO091','NO0A2','NO082']
).sort_values('NUTS3_ID')
hydro_fractions
```

```python
# this could be written in a better way in e.g. a for loop.
eia_stats.NO020 = fraction * hydro_fractions.iloc[0].share_gen
eia_stats.NO060 = fraction * hydro_fractions.iloc[1].share_gen
eia_stats.NO071 = fraction * hydro_fractions.iloc[2].share_gen
eia_stats.NO074 = fraction * hydro_fractions.iloc[3].share_gen
eia_stats.NO081 = fraction * hydro_fractions.iloc[4].share_gen
eia_stats.NO082 = fraction * hydro_fractions.iloc[5].share_gen
eia_stats.NO091 = fraction * hydro_fractions.iloc[6].share_gen
eia_stats.NO092 = fraction * hydro_fractions.iloc[7].share_gen
eia_stats.NO0A1 = fraction * hydro_fractions.iloc[8].share_gen
eia_stats.NO0A2 = fraction * hydro_fractions.iloc[9].share_gen
eia_stats.NO0A3 = fraction * hydro_fractions.iloc[10].share_gen
```
